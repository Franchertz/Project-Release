## Choose and Train Classification Model

### Subtask:
Encode the categorical features in `X_train` and `X_test` using one-hot encoding, ensuring consistency between the training and testing sets.


**Action**:
To encode the categorical features using one-hot encoding, I will import `OneHotEncoder`, create an instance, identify categorical columns in the training and test sets, fit and transform the data, and then combine the encoded features with the numerical features, following the provided instructions.



```
from sklearn.preprocessing import OneHotEncoder

# 3. Identify the categorical columns in X_train and X_test
categorical_features_train = [col for col in categorical_cols if col in X_train.columns]
categorical_features_test = [col for col in categorical_cols if col in X_test.columns]

# Ensure consistency, should be the same list if X_train and X_test come from the same df
# For robustness, we'll use categorical_features_train for both.

# 2. Create an instance of OneHotEncoder
ohe = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

# 4. Fit the OneHotEncoder to the identified categorical columns of X_train
ohe.fit(X_train[categorical_features_train])

# 5. Transform the categorical columns of both X_train and X_test
X_train_encoded_array = ohe.transform(X_train[categorical_features_train])
X_test_encoded_array = ohe.transform(X_test[categorical_features_test])

# 6. Convert the transformed arrays into DataFrames
# Get feature names for the new columns
encoded_feature_names = ohe.get_feature_names_out(categorical_features_train)

X_train_ohe = pd.DataFrame(X_train_encoded_array, columns=encoded_feature_names, index=X_train.index)
X_test_ohe = pd.DataFrame(X_test_encoded_array, columns=encoded_feature_names, index=X_test.index)

# 7. Drop the original categorical columns from X_train and X_test
X_train_numerical = X_train.drop(columns=categorical_features_train)
X_test_numerical = X_test.drop(columns=categorical_features_test)

# 8. Concatenate the one-hot encoded DataFrames with the numerical columns
X_train_encoded = pd.concat([X_train_numerical, X_train_ohe], axis=1)
X_test_encoded = pd.concat([X_test_numerical, X_test_ohe], axis=1)

print("X_train_encoded shape:", X_train_encoded.shape)
print("X_test_encoded shape:", X_test_encoded.shape)
print("X_train_encoded head:\n", X_train_encoded.head())
print("X_test_encoded head:\n", X_test_encoded.head())
```

## Choose and Train Classification Model

### Subtask:
Select an appropriate classification model (e.g., Logistic Regression, Support Vector Machine, or Random Forest Classifier) and train it using your prepared training datasets (X_train_encoded, y_train).


**Action**:
I will select Logistic Regression as a suitable model for binary classification, import it, instantiate it with a `random_state` for reproducibility, and then train it using the preprocessed training data (`X_train_encoded`, `y_train`).



```
from sklearn.linear_model import LogisticRegression

# 1. Choose a classification model: Logistic Regression
# 3. Instantiate the model
# Using solver='liblinear' as it's good for small datasets and handles L1/L2 penalties.
# Set random_state for reproducibility.
model = LogisticRegression(random_state=42, solver='liblinear')

# 4. Train the instantiated model
model.fit(X_train_encoded, y_train)

print("Logistic Regression model trained successfully.")
```

**Action**:
The model has been trained. The next step is to evaluate its performance on the test set by making predictions and then assessing the accuracy.



```
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Make predictions on the test set
y_pred = model.predict(X_test_encoded)

# Evaluate the model's performance
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print(f"Model Accuracy: {accuracy:.4f}")
print("\nClassification Report:")
print(report)
print("\nConfusion Matrix:")
print(conf_matrix)
```

## Visualize Model Performance

### Subtask:
Generate visualizations, such as a ROC curve, precision-recall curve, or confusion matrix heatmap, to provide a clear understanding of the model's predictive capabilities.


**Action**:
The subtask asks to generate several visualizations to assess model performance. I will create a confusion matrix heatmap, an ROC curve with AUC score, and a Precision-Recall curve with average precision score, as specified in the instructions. I will use `seaborn` and `matplotlib` for plotting and `sklearn.metrics` for calculations.



```
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import roc_curve, auc, precision_recall_curve, average_precision_score

# Get predicted probabilities for the positive class (class 1)
y_pred_proba = model.predict_proba(X_test_encoded)[:, 1]

# 1. Plot the Confusion Matrix as a heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=['Predicted 0', 'Predicted 1'], yticklabels=['Actual 0', 'Actual 1'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

# 2. Compute and Plot the ROC Curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()

# 3. Compute and Plot the Precision-Recall Curve
precision, recall, _ = precision_recall_curve(y_test, y_pred_proba)
average_precision = average_precision_score(y_test, y_pred_proba)

plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='purple', lw=2, label=f'Precision-Recall curve (AP = {average_precision:.2f})')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='lower left')
plt.grid(True)
plt.show()
```
